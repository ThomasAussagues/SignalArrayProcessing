```python
import scipy.io as spio
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import matplotlib.pylab as pl
import matplotlib as mpl
mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}']
mpl.rcParams['figure.dpi']=300
mpl.rc('text', usetex=True)

plt.rcParams.update({'font.size': 16})

from tqdm import tqdm
from utils.ASA import get_params,ASA_propagator,ASA_sources
from utils.FFT import custom_fft2,custom_ifft2




```

# 1


```python
config = get_params()
radius = 15*config['lambda']
prop = ASA_propagator(config=config,z=0.08)
U_0 = ASA_sources(config=config,type='square',r=radius)

U = custom_ifft2(custom_fft2(U_0)*prop)
```

```python
plt.pcolormesh(config['ax'],config['ay'],10*np.log10(np.abs(U)),cmap='inferno',shading='auto')
plt.clim(np.min(10*np.log10(np.abs(U)))+40,np.max(10*np.log10(np.abs(U))))
plt.xlabel('Position $x$ (m)')
plt.ylabel('Position $y$ (m)')
plt.title('Wave field amplitude (dB) at depth $z=0.08$ m \n for a squared source, estimated using ASA')
plt.colorbar()
plt.savefig('images/1_wave_field_2D.png',dpi=300)
```

```python
plt.plot(config['ax'],np.abs(U[:,511]),color='blue',linewidth=1)
plt.xlabel('Position $x$ (m)')
plt.ylabel('Amplitude $U$ (dB)')
plt.title('Wave field amplitude (dB) at depth $z=8$ cm and $y=0$ cm \n for a squared source, estimated using ASA')
plt.savefig('imges/1_wave_field_1D.png')
#plt.colorbar()
```

```python
import scipy

def hanning(n):
    m = n/2;
    w = .5*(1 - np.cos(2*np.pi*np.arange(0,m,1))/(n-1))
    print(w)
                    
    return w
                    
hanning(3)
            
```

```python
hanning_window = np.array([[0.25,0.5,0.25],[0.5,1,0.5],[0.25,0.5,0.25]])
print(hanning_window.shape)
```

```python
U_0_mod = scipy.signal.convolve2d(U_0,hanning_window,'same')
```

```python
z_min = 0 #m
z_max = 2 #m
Nz=2000
depth_sampling_period = (z_max-z_min)/Nz

# Depth vector Z

Z = np.arange(z_min,z_max,depth_sampling_period)

# Wave field matrix W (each column <-> wavefield for all x, at y=0 cm and z)

W = np.zeros((config['Nx'],len(Z)),dtype='complex')

for i in tqdm(range(len(Z))) :
    
    prop = ASA_propagator(config=config,z=Z[i])
    U = custom_ifft2(custom_fft2(U_0_mod)*prop)
    W[:,i] = U[:,config['Ny']//2]
    

```

```python
# 2D image of the predicted wavefield for y=0m
```

```python


plt.pcolormesh(Z,config['ax'],10*np.log10(np.abs(W)),cmap='viridis',shading='auto')
plt.clim(np.min(10*np.log10(np.abs(W[-1,:])))+40,np.max(np.log10(np.abs(W[-1,:]))))

near_field_limit = (2*radius)**2/(4*config['lambda'])
far_field_limit = (2*radius)**2/config['lambda']

plt.plot([near_field_limit,near_field_limit],[np.min(config['ax']),np.max(config['ax'])],linestyle='--',color='k')
plt.plot([far_field_limit,far_field_limit],[np.min(config['ax']),np.max(config['ax'])],linestyle='--',color='k')
plt.ylabel('Position $x$ (m)')
plt.xlabel('Depth $z$ (m)')
plt.title('Horizontal wave field amplitude (dB) at depth $z$\n for a squared source, estimated using ASA')
plt.savefig('images/2_wave_field_2D_xz.png',dpi=300)
plt.colorbar()






```

```python
# Plot the pressure for the center of the field for all ranges (i.e. a 1S plot, along the z-axis,
# and starting in origo).

plt.plot(Z,np.abs(W[config['Nx']//2,:]),color='blue',linewidth=0.5)
plt.xlabel('Depth $z$ (m)')
plt.ylabel('Pressure $U$ (dB)')
plt.title('Pressure for the center of the field (dB) at depth $z$ \n for a squared source, estimated using ASA')
plt.savefig('images/2_wave_field_1D_center.png',dpi=300)
#plt.colorbar()


```

```python
#Plot the center cut of the wave field (as in 1.) at the near-field farfield limit. 
#Describe how you defined this limit.



near_field_farfield_limit = 1/2*(far_field_limit-near_field_limit)
prop = ASA_propagator(config=config,z=near_field_farfield_limit)
U = custom_ifft2(custom_fft2(U_0_mod)*prop)

plt.plot(config['ax'],10*np.log10(np.abs(U[:,config['Nx']//2])),color='blue',linewidth=1)
plt.xlabel('Position $x$ (m)')
plt.ylabel('Amplitude $U$ (dB)')
plt.title('Center cut of the wave field ($x$-axis, $y=0$ cm) at the the near-field \n farfield limit for a squared source, estimated using ASA')
plt.savefig('images/2_wave_field_1D_near_far_lim.png',dpi=300)
```

```python
# Plot the center cut of the wave field at a range much larger than the near-field farfield limit.


prop = ASA_propagator(config=config,z=far_field_limit*100)
U = custom_ifft2(custom_fft2(U_0_mod)*prop)

plt.plot(config['ax'],10*np.log10(np.abs(U[:,config['Nx']//2])),color='blue',linewidth=1)
plt.xlabel('Position $x$ (m)')
plt.ylabel('Amplitude $U$ (dB)')
plt.title('Center cut of the wave field ($x$-axis, $y=0$cm) \n far away the farfield limit for a squared source, estimated using ASA')
plt.savefig('images/2_wave_field_1D_far_lim.png',dpi=300)
```
